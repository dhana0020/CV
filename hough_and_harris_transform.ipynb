```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

img = cv2.imread(r'b.jpg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)


edges = cv2.Canny(gray, 40, 200)


lines = cv2.HoughLinesP(edges, 1, np.pi/180, 50, minLineLength=100, maxLineGap=30)


if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(img, (x1, y1), (x2, y2), (0, 255, 0), 1)

plt.subplot(131)
plt.axis('off')
plt.title('Original Image')
plt.imshow(gray,cmap='gray')


plt.subplot(132)
plt.imshow(edges,cmap='gray')
plt.axis('off')
plt.title('Edges')


plt.subplot(133)
plt.imshow(img)
plt.axis('off')
plt.title('Detected Lines')
plt.show()

```

```python

image_copy = cv2.imread(r'h.jpg')

gray = cv2.cvtColor(image_copy, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
dst = cv2.cornerHarris(gray, 3, 5, 0.06)
dst = cv2.dilate(dst,None)


thresh = 0.02*dst.max()

corner_image = np.copy(image_copy)

for j in range(0, dst.shape[0]):
    for i in range(0, dst.shape[1]):
        if(dst[j,i] > thresh):
            # image, center pt, radius, color, thickness
            cv2.circle( corner_image, (i, j), 1, (0,255,0), 1)


plt.subplot(121)
plt.title('original')
plt.axis('off')
plt.imshow(gray,cmap='gray')


plt.subplot(122)
plt.axis('off')
plt.title('detected corners')
plt.imshow(corner_image)

plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread(r'iii3.jpg')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

edges = cv2.Canny(gray, 40, 200)


lines = cv2.HoughLinesP(edges, 1, np.pi/180, 50, minLineLength=100, maxLineGap=30)


if lines is not None:
    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(img, (x1, y1), (x2, y2), (0, 255, 0), 2)




image_copy1 = cv2.imread(r'iii3.jpg')

g1 = cv2.cvtColor(image_copy1, cv2.COLOR_BGR2GRAY)
g1 = np.float32(gray)
dst1 = cv2.cornerHarris(g1, 3, 5, 0.06)
dst1 = cv2.dilate(dst1,None)


thresh1 = 0.003*dst1.max()

corner_image1 = np.copy(image_copy1)

for j in range(0, dst1.shape[0]):
    for i in range(0, dst1.shape[1]):
        if(dst1[j,i] > thresh1):
            # image, center pt, radius, color, thickness
            cv2.circle( corner_image1, (i, j), 1, (255,0,0), 3)


plt.subplot(221)
plt.axis('off')
plt.title('original ')
plt.imshow(gray,cmap='gray')


plt.subplot(222)
plt.axis('off')
plt.title('line ')
plt.imshow(img)



plt.subplot(223)
plt.axis('off')
plt.title('line ')
plt.imshow(corner_image1)


s = cv2.addWeighted(corner_image1, 0.7, img, 0.5, 0)

plt.subplot(224)
plt.title('segmented (line + corner) ')
plt.axis('off')
plt.imshow(s)
plt.show()
```
