```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
d=cv2.imread(r"C:\Users\hanal\Desktop\360_F_266724172_Iy8gdKgMa7XmrhYYxLCxyhx6J7070Pr8.jpg")
d
```

```python
plt.title('original')

plt.imshow(d)
```

```python
gr=cv2.cvtColor(d,cv2.COLOR_BGR2GRAY)
gr1=cv2.resize(gr, (500, 500))
plt.title("GRAY image after resizing")
plt.imshow(gr1,cmap='grey')

```

```python
(r, c) = d.shape[0:2]
for i in range(r):
    for j in range(c):
        d[i, j] = sum(d[i, j]) * 0.33
plt.title('GRAY Image WIHOUT FUNCTION without resize')

plt.imshow(d)
```

```python
d=cv2.imread(r"C:\Users\hanal\Desktop\360_F_266724172_Iy8gdKgMa7XmrhYYxLCxyhx6J7070Pr8.jpg")

import numpy as np
eq1 = cv2.equalizeHist(d[:,:,0])
eq2 = cv2.equalizeHist(d[:,:,1])
eq3 = cv2.equalizeHist(d[:,:,2])

feq = np.stack((eq1,eq2,eq3), axis=2)
```

```python
plt.subplot(121)
plt.title("RGB image")
plt.axis('off')
plt.imshow(d)
plt.subplot(122)
plt.title("equalised image")
plt.axis('off')
plt.imshow(feq)
```

```python
import matplotlib.pyplot as plt
r = cv2.calcHist([d], [0], None, [256], [0, 256]) 
g = cv2.calcHist([d], [1], None, [256], [0, 256]) 
b = cv2.calcHist([d], [2], None, [256], [0, 256]) 
plt.figure(figsize=(20,5))
# combined histogram 
plt.title("RGB histogram hist vs equalised hist")
plt.subplot(1,4,1)
plt.hist(r, color="red") 
plt.subplot(1,4,2)
plt.hist(g, color="green") 
plt.subplot(1,4,3)
plt.hist(b, color="blue") 
plt.subplot(1,4,4)
f=feq.flatten()
plt.hist(f)
plt.show() 
```

```python
import matplotlib.pyplot as plt
eq = cv2.equalizeHist(gr1)
plt.subplot(121)
plt.title("gray image")
plt.axis('off')
plt.imshow(gr1)
plt.subplot(122)
plt.title("equalised image")
plt.axis('off')
eq1=cv2.resize(eq, (500, 500))
plt.imshow(eq1)
```

```python

plt.figure(figsize=(10,5))
plt.title("gray hist vs  gray equalised hist")
plt.subplot(1,2,1)
plt.hist(gr)
plt.subplot(1,2,2)
plt.hist(eq)
```
