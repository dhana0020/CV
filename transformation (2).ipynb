rotation

```python
from math import radians
d=cv2.imread(r"C:\Users\student\Desktop\i1.jpg")
a=radians(45)
fi2 = np.zeros_like(d)

xc=r // 2
yc=c // 2
for i in range(r):
    for j in range (c):
        x=i-xc
        y=j-yc
        xn=int((x*np.cos(a))-(y*np.sin(a)))+xc
        yn=int((x*np.sin(a))+(y*np.cos(a)))+yc
        if 0<=xn < r and 0<=yn <c:
            fi2[xn,yn]=d[i,j]
plt.subplot(121)
plt.title('original')

plt.imshow(d)   
plt.subplot(122)
plt.title('rotated')
plt.imshow(fi2)
```
